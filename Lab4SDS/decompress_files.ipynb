```python
import os
import pefile

from pathlib import Path
```

```python
def find_compress_files(folder_path):
    folder_path = Path(folder_path)
    files = [file for file in folder_path.iterdir() if file.is_file()]
    compress_folder_paths = []

    for num, file_path in enumerate(files):
        if '.' in file_path.name:
            continue
        pe = pefile.PE(file_path)
        for section in pe.sections:
            if "UPX" in str(section.Name):
                print(f"UPX section found in num: {num} with file path: {file_path}")
                compress_folder_paths.append(file_path)
                break

    return compress_folder_paths
```

```python
compress_folder_paths = find_compress_files("MALWR")
```

```python
print(len(compress_folder_paths))
```

```python
for file_path in compress_folder_paths:
    os.system(f"upx -d {file_path}")
```

```python
compress_folder_paths_after = find_compress_files("MALWR")
```
